## Chuẩn bị

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as plt_sns

plt_sns.set_theme(style="whitegrid")

DATA_DIR = "processed_data/"

def load_all_champions():
    data_list = []
    for filename in os.listdir(DATA_DIR):
        if filename.endswith(".json"):
            filepath = os.path.join(DATA_DIR, filename)
            with open(filepath, 'r', encoding='utf-8') as f:
                champ_data = json.load(f)
                # Lưu thêm tên file làm ID tướng
                champ_data['champion_id'] = filename.replace('.json', '')
                data_list.append(champ_data)
    return data_list

champions = load_all_champions()
print(f"Đã load thành công {len(champions)} tướng!")
```

## Phân tích Cấu trúc Infobox (Metadata EDA)
- Phân bố tướng theo vùng đất
- Phân bố thuộc tính quan trọng

```python
def eda_metadata(champions_data):
    print("\n--- 1. METADATA EDA (INFOBOX) ---")
    
    # 1.1 Chuẩn bị dữ liệu dạng bảng (DataFrame)
    rows = []
    for champ in champions_data:
        info = champ.get('infobox', {})
        rows.append({
            'Champion': champ['champion_id'],
            'Has_Family': 1 if 'Family' in info else 0,
            'Has_Related': 1 if 'Related character' in info else 0,
            'Has_Weapon': 1 if 'Weapon(s)' in info or 'Weapon' in info else 0,
            'Has_Species': 1 if 'Species' in info else 0,
            # Lấy vùng đất đầu tiên nếu có
            'Primary_Region': info.get('Region(s)', info.get('Region', ['Unknown']))[0]
        })
        
    df = pd.DataFrame(rows)
    
    # 1.2 Thống kê độ khuyết thiếu (Missing Values)
    print("\nTỷ lệ Tồn tại của các trường dữ liệu quan trọng (%):")
    missing_stats = df[['Has_Family', 'Has_Related', 'Has_Weapon', 'Has_Species']].mean() * 100
    print(missing_stats.round(2))
    
    # 1.3 Vẽ biểu đồ Phân bố Vùng đất (Categorical Distribution)
    plt.figure(figsize=(12, 6))
    region_counts = df['Primary_Region'].value_counts()
    
    # Lọc bỏ các giá trị quá nhỏ để biểu đồ đỡ rối (chỉ lấy Top 15)
    top_regions = region_counts.head(30)
    
    plt_sns.barplot(x=top_regions.values, y=top_regions.index, hue=top_regions.index, legend=False)
    plt.title("Phân bố Tướng theo Vùng đất (Top 15)", fontsize=14, fontweight='bold')
    plt.xlabel("Số lượng tướng")
    plt.ylabel("Vùng đất (Region)")
    plt.tight_layout()
    plt.show()

# Thực thi
eda_metadata(champions)
```

## Phân tích Mạng lưới Thực thể sơ bộ (Network EDA)

```python
import networkx as nx

def eda_network(champions_data):
    print("\n--- 2. NETWORK EDA (RELATED CHARACTERS) ---")
    
    # 2.1 Khởi tạo Đồ thị có hướng (Directed Graph)
    G = nx.DiGraph()
    
    for champ in champions_data:
        source_node = champ.get('infobox', {}).get('Title', [champ['champion_id']])[0]
        related_chars = champ.get('infobox', {}).get('Related character', [])
        
        # Thêm node gốc để đảm bảo tướng không có quan hệ vẫn tồn tại trên đồ thị
        G.add_node(source_node)
        
        for target_node in related_chars:
            G.add_edge(source_node, target_node)
            
    print(f"Tổng số Nút (Thực thể): {G.number_of_nodes()}")
    print(f"Tổng số Cạnh (Mối quan hệ): {G.number_of_edges()}")
    
    # 2.2 Tính toán Mật độ (Degree Centrality)
    # Dùng In-Degree (Số người nhắc đến họ) + Out-Degree (Số người họ nhắc đến)
    degrees = dict(G.degree())
    
    # Sắp xếp để lấy Top 10 Hubs (Nhiều quan hệ nhất)
    sorted_hubs = sorted(degrees.items(), key=lambda item: item[1], reverse=True)
    
    print("\n👑 Top 5 Tướng 'Ngoại giao' (Hubs - Nhiều quan hệ nhất):")
    for name, degree in sorted_hubs[:5]:
        print(f" - {name}: {degree} mối quan hệ")
        
    # Tìm những kẻ cô độc (Degree = 0)
    isolated = [name for name, degree in degrees.items() if degree == 0]
    print(f"\n🐺 Có {len(isolated)} thực thể 'Cô độc' (Không liên kết với ai).")
    if isolated:
        print(f" Ví dụ: {', '.join(isolated[:5])}...")

# Thực thi
eda_network(champions)
```

### Vấn đề Định danh

```python
import json

def eda_entity_resolution(champions_data):
    print("\n--- 4. ENTITY RESOLUTION (ALIAS MAPPING) ---")
    
    alias_dict = {}
    alias_stats = []
    
    for champ in champions_data:
        canonical_name = champ['champion_id'] # Tên gốc chuẩn (VD: "Lux")
        info = champ.get('infobox', {})
        
        # Tạo một tập hợp (set) để chứa tất cả các tên gọi, tránh trùng lặp
        all_names = set()
        
        # 1. Quét các trường chứa tên gọi trong Infobox
        fields_to_check = ['Title', 'Real name', 'Alias', 'Nickname']
        
        for field in fields_to_check:
            if field in info:
                # Dữ liệu từ Infobox đang là dạng list các chuỗi
                for name in info[field]:
                    # Dọn dẹp: Đưa về chữ thường để dễ so sánh (case-insensitive)
                    clean_name = name.strip().lower()
                    if clean_name:
                        all_names.add(clean_name)
        
        # 2. Thêm tên gốc vào danh sách nếu chưa có
        all_names.add(canonical_name.lower())
        
        # 3. Lưu số liệu thống kê
        alias_stats.append({
            'Champion': canonical_name,
            'Alias_Count': len(all_names) - 1 # Trừ đi cái tên gốc
        })
        
        # 4. Xây dựng bộ Từ điển (Mapping)
        for name in all_names:
            # Nếu tên này đã bị ai đó chiếm dụng (Trường hợp trùng bí danh hiếm hoi)
            if name in alias_dict and alias_dict[name] != canonical_name:
                print(f"⚠️ CẢNH BÁO XUNG ĐỘT: '{name}' trỏ tới cả {alias_dict[name]} và {canonical_name}")
            else:
                alias_dict[name] = canonical_name

    # --- IN KẾT QUẢ THỐNG KÊ ---
    df_aliases = pd.DataFrame(alias_stats)
    print(f"Tổng số 'Bí danh' / Tên gọi thu thập được: {len(alias_dict)}")
    
    print("\n🎭 Top 5 Tướng có nhiều tên gọi nhất:")
    top_aliases = df_aliases.sort_values(by='Alias_Count', ascending=False).head(5)
    for _, row in top_aliases.iterrows():
        print(f" - {row['Champion']}: {row['Alias_Count']} tên gọi")
        
    # --- XUẤT FILE BẢO BỐI ---
    output_path = 'alias_mapping.json'
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(alias_dict, f, indent=4, ensure_ascii=False)
        
    print(f"\n✅ Đã xuất bảng Mapping ra file: {output_path}")
    print("Mẫu dữ liệu Mapping sinh ra:")
    
    # In thử vài dòng của Lux để kiểm chứng
    lux_samples = {k: v for k, v in alias_dict.items() if v == 'Lux'}
    for k, v in list(lux_samples.items())[:5]:
        print(f"   '{k}' -> sẽ được quy chuẩn thành Node: ({v})")

# Thực thi
eda_entity_resolution(champions)
```

## Phân tích Đặc trưng Văn bản
- Phân phối Độ dài Tiểu sử
- Đồng nhất Cấu trúc Mục lục

```python
def extract_text_and_sections(toc_items, section_counts, current_words=0):
    """Hàm đệ quy để gom chữ và đếm mục lục"""
    words = current_words
    for item in toc_items:
        title = item.get('title', '').strip()
        text = item.get('textContent', '')
        
        if title:
            section_counts[title] = section_counts.get(title, 0) + 1
            
        if text:
            words += len(text.split()) # Đếm số chữ (Word count)
            
        if item.get('children'):
            words = extract_text_and_sections(item['children'], section_counts, words)
            
    return words

def eda_text(champions_data):
    print("\n--- 3. TEXTUAL EDA (MAIN CONTENT) ---")
    
    word_counts = []
    section_frequency = {}
    
    for champ in champions_data:
        toc = champ.get('mainContent', [])
        total_words = extract_text_and_sections(toc, section_frequency)
        word_counts.append({
            'Champion': champ['champion_id'],
            'WordCount': total_words
        })
        
    df_text = pd.DataFrame(word_counts)
    
    # Lọc ra những tướng không có chữ nào (bị lỗi parse hoặc Wiki rỗng)
    df_valid = df_text[df_text['WordCount'] > 0]
    
    print(f"\nThống kê Độ dài Tiểu sử (Số từ):")
    print(df_valid['WordCount'].describe().round(0))
    
    # 3.1 Vẽ biểu đồ Phân phối độ dài tiểu sử
    plt.figure(figsize=(10, 5))
    plt_sns.histplot(df_valid['WordCount'], bins=30, kde=True, color='coral')
    plt.title("Phân phối Độ dài Tiểu sử Tướng (Word Count)", fontsize=14, fontweight='bold')
    plt.xlabel("Số lượng từ (Words)")
    plt.ylabel("Số lượng Tướng")
    
    # Vẽ đường trung bình để dễ nhìn
    plt.axvline(df_valid['WordCount'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    # 3.2 Top 10 Mục lục phổ biến nhất (TOC Consistency)
    print("\n📑 Top 10 Mục lục (Section Titles) xuất hiện nhiều nhất:")
    sorted_sections = sorted(section_frequency.items(), key=lambda x: x[1], reverse=True)
    for title, count in sorted_sections[:10]:
        # Tính % xuất hiện trên tổng số tướng
        percentage = (count / len(champions_data)) * 100
        print(f" - {title}: {count} lần ({percentage:.1f}%)")

# Thực thi
eda_text(champions)
```
